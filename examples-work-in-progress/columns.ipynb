```python
import matplotlib.pyplot as plt
import numpy as np

col1, col2, col3 = Columns(3)

with col1:
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x)
    plt.figure(figsize=(3,2))
    plt.plot(x, y)
    plt.title('Sine Wave')
    plt.tight_layout()
    plt.show()

with col2:
    x = np.linspace(0, 2*np.pi, 100)
    y = np.cos(x)
    plt.figure(figsize=(3,2))
    plt.plot(x, y, color='orange')
    plt.title('Cosine Wave')
    plt.tight_layout()
    plt.show()

with col3:
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(x) * np.cos(x)
    plt.figure(figsize=(3,2))
    plt.plot(x, y, color='green')
    plt.title('Sin(x)*Cos(x)')
    plt.tight_layout()
    plt.show()
```
